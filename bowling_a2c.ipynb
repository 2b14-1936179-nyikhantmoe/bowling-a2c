### Baseline Model 
No changes to the parameters

```
import gym

from stable_baselines3 import A2C
from stable_baselines3.a2c import MlpPolicy
from stable_baselines3.common.env_util import make_vec_env

# Parallel environments
env = make_vec_env('Bowling-v0', n_envs=4)

model = A2C(MlpPolicy, env, verbose=1)
model.learn(total_timesteps=25000)
model.save("a2c_bowl-base")

del model # remove to demonstrate saving and loading

model = A2C.load("a2c_bowl-base")
```

```
import gym
import gnwrapper
env = gym.make('Bowling-v0')
env = gnwrapper.Monitor(env, directory="./bowl-base", force=True)
```

```
obs = env.reset()
total, miss, score = 0, 0, 0
while True:
  env.render()
  action, state = model.predict(obs, deterministic=False)
  obs, reward, done, info = env.step(action)
  total += 1
  if reward > 0:
    score += reward
    print(reward, ',', done)
  if reward == 0:
    miss += 1         
  if done:
    print('Total:', total, '| Misses:', miss, '| Final score:', score)
    break;
env.close()
```

```
import gym

from stable_baselines3 import A2C
from stable_baselines3.a2c import MlpPolicy
from stable_baselines3.common.env_util import make_vec_env

# Parallel environments
env = make_vec_env('Bowling-v0', n_envs=8)
modelMLP2 = A2C(MlpPolicy, env, verbose=1, learning_rate=0.0005, n_steps=5, gamma=0.99, gae_lambda=1.0, 
               ent_coef=0.0, vf_coef=0.5, max_grad_norm=0.5, rms_prop_eps=1e-05, use_rms_prop=True, 
               use_sde=False, sde_sample_freq=-1, normalize_advantage=False, tensorboard_log=None, 
               create_eval_env=False, policy_kwargs=None, seed=1, device='auto', _init_setup_model=True)
# changes to the timesteps
modelMLP2.learn(total_timesteps=50000)
modelMLP2.save("a2c_bowling-MLP-2")

del modelMLP2 # remove to demonstrate saving and loading

modelMLP2 = A2C.load("a2c_bowling-MLP-2")
```

```
import gym
import gnwrapper
env = gym.make('Bowling-v0')
env = gnwrapper.Monitor(env, directory="./bowl-MLP-2", force=True)
```

```
obs = env.reset()
total, miss, score = 0, 0, 0
while True:
  env.render()
  action, state = modelMLP2.predict(obs, deterministic=False)
  obs, reward, done, info = env.step(action)
  total += 1
  if reward > 0:
    score += reward
    print(reward, ',', done)
  if reward == 0:
    miss += 1         
  if done:
    print('Total:', total, '| Misses:', miss, '| Final score:', score)
    break;
env.close()
```

The scores improved after changing the timesteps.
